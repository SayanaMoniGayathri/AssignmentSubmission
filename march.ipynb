```python
def calculate_stats(df, label):
    wins = df.groupby(['Season', 'WTeamID']).size().reset_index(name=f'{label}_Wins')
    wins.rename(columns={'WTeamID': 'TeamID'}, inplace=True)

    losses = df.groupby(['Season', 'LTeamID']).size().reset_index(name=f'{label}_Losses')
    losses.rename(columns={'LTeamID': 'TeamID'}, inplace=True)

    stats = pd.merge(wins, losses, on=['Season', 'TeamID'], how='outer').fillna(0)

    stats[f'{label}_WinPercentage'] = stats[f'{label}_Wins'] / (stats[f'{label}_Wins'] + stats[f'{label}_Losses'])

    return stats

regular_season_stats = calculate_stats(dfs['MRegularSeasonCompactResults'], 'RegularSeason')
tourney_stats = calculate_stats(dfs['MNCAATourneyCompactResults'], 'Tourney')
secondary_tourney_stats = calculate_stats(dfs['MSecondaryTourneyCompactResults'], 'SecondaryTourney')

all_stats = pd.merge(regular_season_stats, tourney_stats, on=['Season', 'TeamID'], how='outer')
all_stats = pd.merge(all_stats, secondary_tourney_stats, on=['Season', 'TeamID'], how='outer')

all_stats
```

```python
def calculate_goals_accuracy(df):
    df['WFGP'] = df['WFGM'] / df['WFGA']
    df['LFGP'] = df['LFGM'] / df['LFGA']

    agg_win_metrics = df.groupby('WTeamID').agg(
        WFGP_mean=('WFGP', 'mean')
    ).reset_index()

    # Aggregate losing team metrics
    agg_lose_metrics = df.groupby('LTeamID').agg(
        LFGP_mean=('LFGP', 'mean')
    ).reset_index()

    # Rename columns for merging
    agg_win_metrics.rename(columns={'WTeamID': 'TeamID', 'WFGP_mean': 'FGP_mean'}, inplace=True)
    agg_lose_metrics.rename(columns={'LTeamID': 'TeamID', 'LFGP_mean': 'FGP_mean'}, inplace=True)

    # Concatenate win and lose metrics
    combined_metrics = pd.concat([agg_win_metrics, agg_lose_metrics]).groupby('TeamID', as_index=False).mean()

    return combined_metrics

# Calculate goals accuracy for the provided DataFrame
temp_df = dfs['MRegularSeasonDetailedResults']
goals_accuracy_df = calculate_goals_accuracy(temp_df)
goals_accuracy_df.head()
```
